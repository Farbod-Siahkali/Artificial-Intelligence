1) IDF-TF, also known as Inverse Document Frequency - Term Frequency, is a technique used in natural language processing (NLP) to quantify the importance of a term in a document or a collection of documents. It combines two measures: Term Frequency (TF) and Inverse Document Frequency (IDF).

* Term Frequency (TF) calculates the frequency of a term in a document. It measures how often a term appears in a document and provides a local importance score for the term within that document.

* Inverse Document Frequency (IDF) measures the rarity of a term across the entire document collection. It assigns a global importance score to the term based on how common or rare it is in the collection. Terms that appear frequently across many documents receive a lower IDF score, while terms that appear in a few documents receive a higher IDF score.

The IDF-TF score is obtained by multiplying the Term Frequency (TF) with the Inverse Document Frequency (IDF) for each term in a document. This helps to identify terms that are both frequently occurring in the document and relatively rare in the entire collection, thus capturing their significance.

2) Stemming and Lemmatization are techniques used in NLP for word normalization. They are used to transform words into their base or root form to reduce inflectional variations and improve text analysis.

* Stemming reduces words to their base or root form by removing suffixes or prefixes. It follows simple rules and heuristics to chop off affixes, but it may not always result in a valid word. For example, stemming the word "running" would give "run."

* Lemmatization, on the other hand, considers the context and part of speech of a word to convert it into its base form, known as a lemma. It uses linguistic knowledge and a dictionary to ensure valid words. For example, lemmatizing the word "running" would give "run" as the lemma.

In the work report, replacing words using stemming or lemmatization aims to reduce word variations and bring related words to a common base form. This helps in improving the accuracy of text analysis tasks such as text classification or information retrieval.

3) Ensuring that each feature's value is between 0 and 1 before feeding it into a neural network is important for several reasons:

* Gradient-based optimization algorithms, such as backpropagation, work well when the input features are on a similar scale. If features have different scales, some features may dominate the learning process, leading to slower convergence or suboptimal results.

*Activation functions in neural networks, such as sigmoid or tanh, are sensitive to the input range. If inputs are too large or too small, these activation functions can saturate, causing gradients to become very small and leading to the vanishing gradient problem. This hinders the learning process.

* Regularization techniques, like L1 or L2 regularization, also assume that features are on a similar scale. If features have different scales, the regularization penalties may disproportionately affect certain features, leading to biased learning.

If the condition of having values between 0 and 1 is not met, several problems may arise:

* Slow convergence: It may take longer for the network to converge to an optimal solution or it may fail to converge at all.

*Unstable gradients: The gradients calculated during backpropagation can become very large or very small, making the learning process unstable.

* Loss of information: If features have significantly different scales, the network may not effectively capture the importance of each feature, leading to a loss of valuable information.

4) If all the initial weights of a network were set to zero and the network is trained, the resulting consequences would be that all neurons in the network would receive the same gradient during backpropagation. As a result, all neurons would update their weights in the same way and become symmetric in terms of their learned features.

This symmetry breaking problem can lead to several issues:

* Limited representational capacity: The network may not be able to learn complex and diverse patterns as all neurons will have the same learned features.

* No discrimination power: Neurons with the same weights will produce identical outputs for any given input, rendering the network unable to differentiate between different classes or patterns.

* Stagnation during training: With identical weights, the network's parameters will not update differently based on the input data, causing the training process to stagnate without achieving meaningful progress.

To overcome these issues, it is important to initialize the weights of a neural network with small random values, enabling each neuron to learn unique features and improve the network's learning capacity.

5) Tanh and Sigmoid functions are generally not considered suitable choices for hidden layers in neural networks due to the following reasons:

* Vanishing gradients: Both Tanh and Sigmoid functions saturate towards the extremes, where their derivatives become very small. This leads to the vanishing gradient problem, especially in deep networks, where gradients diminish rapidly during backpropagation and hinder the learning of lower layers.

* Output range: Tanh and Sigmoid functions have limited output ranges (-1 to 1 for Tanh and 0 to 1 for Sigmoid). This can cause activation values to be concentrated around the extremes, leading to slow convergence and the vanishing gradient problem.

* Non-zero centered output: The output of Tanh and Sigmoid functions is non-zero centered, which can introduce biases in the network. This may slow down learning or make it difficult for the network to find an optimal solution.

Instead, activation functions like ReLU (Rectified Linear Unit) or its variants, such as Leaky ReLU or Parametric ReLU, are often preferred for hidden layers. They alleviate the vanishing gradient problem, allow faster convergence, and promote sparsity in activation values, leading to improved training speed and generalization.

6) Leaky ReLU is an improved version of the ReLU (Rectified Linear Unit) activation function. The key difference lies in the way negative values are handled:

* ReLU sets all negative values to zero, effectively eliminating the negative part of the input. However, this can lead to dead neurons (neurons that never activate) during training if their biases are set such that they always produce negative outputs.

* Leaky ReLU introduces a small slope (typically a small constant, like 0.01) for negative values instead of setting them to zero. This allows a small gradient to flow through the neurons even for negative inputs, preventing the issue of dead neurons.

The superiority of Leaky ReLU over ReLU lies in its ability to mitigate the dead neuron problem, which can occur when using ReLU. Leaky ReLU helps to address this issue and has shown improved performance in certain scenarios, providing more robust and stable training.

7) Batches are used in the training process for several reasons:

* Efficiency: Training on individual data points can be computationally expensive. By using batches, parallel processing capabilities of modern hardware, such as GPUs, can be leveraged to perform computations on multiple data points simultaneously, leading to faster training times.

* Generalization: Training on batches allows the network to see different examples within each batch, leading to a more generalized model. It helps the network learn from a diverse set of examples, reducing overfitting to individual data points.

* Smoother gradient estimation: The gradients calculated during backpropagation are averaged over the examples in a batch, providing a smoother estimation of the true gradient. This can lead to more stable and consistent updates to the network's parameters.

Advantages of small batch sizes:

* More frequent weight updates: Smaller batch sizes result in more frequent weight updates, allowing the network to adapt faster to the training data.

* Less memory usage: Smaller batch sizes require less memory to store the intermediate activations and gradients during training, making it more memory-efficient, especially for larger models.

Disadvantages of small batch sizes:

* Noisier gradient estimates: Smaller batch sizes provide noisier estimates of the true gradient, as the gradients are computed based on a subset of the data. This can introduce instability during training and hinder convergence.

* Slower convergence: Smaller batch sizes may require more iterations to converge compared to larger batch sizes. This can result in longer training times.

Advantages of large batch sizes:

* More accurate gradient estimates: Larger batch sizes provide more accurate estimates of the true gradient, as they are computed based on a larger amount of data. This can lead to more stable and reliable updates to the network's parameters.

* Faster convergence: Large batch sizes often lead to faster convergence since each weight update is based on a larger sample of the data, allowing the network to learn more efficiently.

Disadvantages of large batch sizes:

* Higher memory requirements: Larger batch sizes require more memory to store the intermediate activations and gradients during training. This can be a limitation, especially for models with limited memory resources.

* Potential for overfitting: Large batch sizes may reduce the generalization ability of the network, as the model relies heavily on a subset of the data during each weight update. This can lead to overfitting if the batch is not representative of the overall dataset.

The choice of batch size depends on various factors such as the size of the dataset, computational resources, and the specific problem at hand. It is often determined through experimentation and balancing the trade-offs between efficiency, convergence speed, and generalization.

Here you need to code your implementation of the [ReLU](https://en.wikipedia.org/wiki/Rectifier_(neural_networks)) activation and the [Sigmoid](https://en.wikipedia.org/wiki/Sigmoid_function).

```python
class MyReLU(object):
    def forward(self, x):
        self.x = x
        return np.maximum(0, x)
    
    def backward(self, grad_output):
        return grad_output * (self.x > 0)
    
    def step(self, learning_rate):
        pass

class MySigmoid(object):
    def forward(self, x):
        self.x = x
        return 1 / (1 + np.exp(-x))
    
    def backward(self, grad_output):
        output = self.forward(self.x)
        return grad_output * (1 - output) * output
    
    def step(self, learning_rate):
        pass
```

A bit more complicated, you need now to implement your linear layer i.e. multiplication by a matrix W and summing with a bias b.

```python
class MyLinear(object):
    def __init__(self, n_input, n_output):
        self.W = np.random.randn(n_output, n_input)
        self.b = np.random.randn(n_output)

    def forward(self, x):
        self.x = x
        return np.dot(self.W, x) + self.b

    def backward(self, grad_output):
        self.grad_W = np.outer(grad_output, self.x)
        self.grad_b = grad_output
        return np.dot(self.W.T, grad_output)
    
    def step(self, learning_rate):
        self.W -= learning_rate * self.grad_W
        self.b -= learning_rate * self.grad_b
```

Code your network. Note that with a Sigmoid layer, you should use the Binary Cross Entropy (BCE) loss.

```python
class Sequential(object):
    def __init__(self):
        self.layers = []
        
    def add(self, layer):
        self.layers.append(layer)
  
    def forward(self, x):
        for layer in self.layers:
            x = layer.forward(x)
        return x
    
    def compute_loss(self, out, label):
        eps = 1e-10
        self.loss_grad = -((label / (out + eps)) - ((1 - label) / (1 - out + eps)))
        return -(label * np.log(out + eps) + (1 - label) * np.log(1 - out + eps))

    def backward(self):
        grad = self.loss_grad
        for layer in reversed(self.layers):
            grad = layer.backward(grad)
    
    def step(self, learning_rate):
        for layer in self.layers:
            layer.step(learning_rate)
            
```

```python
# Define your network architecture
net = Sequential()
net.add(MyLinear(2, 16))
net.add(MyReLU())
net.add(MyLinear(16, 1))
net.add(MySigmoid())
```

## 3.2. MLP in torch

```python
# Define your data loader here (specify your batch_size)
# Convert X_train and y_train to torch.Tensor
batch_size = 64
train_dataset = TensorDataset(torch.Tensor(X_train), torch.Tensor(y_train))
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

test_dataset = TensorDataset(torch.Tensor(X_test), torch.Tensor(y_test))
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

Define your network using `MyLinear_mod`, `MyReLU_mod` and [`nn.Sigmoid`](https://pytorch.org/docs/stable/nn.html#sigmoid)

```python
# Define your network using nn.Sequential
# Use MyLinear_mod, MyReLU_mod, and nn.Sigmoid (from PyTorch)
class MyLinear_mod(nn.Module):
    def __init__(self, n_input, n_output):
        super(MyLinear_mod, self).__init__()
        self.A = nn.Parameter(torch.randn(n_input, n_output))
        self.b = nn.Parameter(torch.randn(n_output))

    def forward(self, x):
        return torch.matmul(x, self.A) + self.b

class MyReLU_mod(nn.Module):
    def __init__(self):
        super(MyReLU_mod, self).__init__()

    def forward(self, x):
        return torch.relu(x)
```

```python
# Create network using nn.Sequential
TorchNet = nn.Sequential(
    MyLinear_mod(5000, 500),
    MyReLU_mod(),
    MyLinear_mod(500, 100),
    MyReLU_mod(),
    MyLinear_mod(100, 1),
    nn.Sigmoid()
)
```

## Training the Network

```python
# Define your optimizer
optimizer = torch.optim.SGD(TorchNet.parameters(), lr=0.02)

# Define the loss function
criterion = nn.BCELoss()

# Define the number of epochs
epochs = 10

# Training loop
train_losses = []
test_losses = []

for epoch in range(epochs):
    temp = []
    TorchNet.train()
    for batch in train_dataloader:
          # Perform a forward pass on the batch
          x, y = batch
          output = TorchNet(x)
          # Compute the loss according to your output and the label
          loss = criterion(output, y.unsqueeze(1))
          # Zero the gradients
          optimizer.zero_grad()
          # Backward pass
          loss.backward()
          # Gradient step
          optimizer.step()
          # Append the current loss value to the list
          temp.append(loss.item())

    train_losses.append(sum(temp)/len(temp))
    TorchNet.eval()
    temp = []
    for batch in test_dataloader:
          # Perform a forward pass on the batch
          x, y = batch
          output = TorchNet(x)
          # Compute the loss according to your output and the label
          loss = criterion(output, y.unsqueeze(1))
          # Append the current loss value to the list
          temp.append(loss.item())

    test_losses.append(sum(temp)/len(temp))
    print('Epoch: ', epoch+1)
    print('Train Loss: ', train_losses[-1])
    print('Test Loss: ', test_losses[-1],'\n')
```

```python
# Plot the loss curve
plt.plot(train_losses, c='r', label='Train')
plt.plot(test_losses, c='k', label='Test')
plt.grid()
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## Evaluation

```python
# Print the classification report

# Evaluate the network on the test data
TorchNet.eval()
y_pred = []
for batch in test_dataloader:
    x = batch[0]
    output = TorchNet(x)
    predicted_labels = (output > 0.5).float()
    y_pred.extend(predicted_labels.tolist())

# Convert y_pred to numpy array
y_pred = np.array(y_pred)

# Print the classification report
target_names = ['negative', 'positive']
print(classification_report(y_test, y_pred, target_names=target_names))
```
